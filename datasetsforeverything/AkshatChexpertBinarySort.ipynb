```python
import pandas
import os
```

```python
"""

Set to file path where CheXper-v1.0-small is located

For example, if your directory looks like '/Users/tperera/documents/XRayData/CheXper-v1.0-small', then set the path to '/Users/tperera/documents/XRayData'

"""

PATH_TO_CHEXPERT_DOWNLOADS = "/Users/tperera/documents/XRayData"

if not os.path.exists(PATH_TO_CHEXPERT_DOWNLOADS + '/train'):
    os.makedirs(PATH_TO_CHEXPERT_DOWNLOADS + '/train/PA/abnormal')
    os.makedirs(PATH_TO_CHEXPERT_DOWNLOADS + '/train/PA/normal')
    os.makedirs(PATH_TO_CHEXPERT_DOWNLOADS + '/train/AP/abnormal')
    os.makedirs(PATH_TO_CHEXPERT_DOWNLOADS + '/train/AP/normal')
    os.makedirs(PATH_TO_CHEXPERT_DOWNLOADS + '/train/L/abnormal')
    os.makedirs(PATH_TO_CHEXPERT_DOWNLOADS + '/train/L/normal')

PATH_TO_NORMAL_TRAIN_DIRECTORY_PA = os.path.join(PATH_TO_CHEXPERT_DOWNLOADS + '/train/PA/normal')
PATH_TO_ABNORMAL_TRAIN_DIRECTORY_PA = os.path.join(PATH_TO_CHEXPERT_DOWNLOADS + '/train/PA/abnormal')
PATH_TO_NORMAL_TRAIN_DIRECTORY_AP = os.path.join(PATH_TO_CHEXPERT_DOWNLOADS + '/train/AP/normal')
PATH_TO_ABNORMAL_TRAIN_DIRECTORY_AP = os.path.join(PATH_TO_CHEXPERT_DOWNLOADS + '/train/AP/abnormal')
PATH_TO_NORMAL_TRAIN_DIRECTORY_L = os.path.join(PATH_TO_CHEXPERT_DOWNLOADS + '/train/L/normal')
PATH_TO_ABNORMAL_TRAIN_DIRECTORY_L = os.path.join(PATH_TO_CHEXPERT_DOWNLOADS + '/train/L/abnormal')
```

```python
# read in chexpert_train.csv
chexpert_train = pandas.read_csv('train.csv')

chexpert_train
```

```python
# remove people with support devices

chexpert_train.drop(chexpert_train[chexpert_train['Support Devices'] == 1].index, inplace=True)

chexpert_train
```

```python
# remove children (under 18)
chexpert_train.drop(chexpert_train[chexpert_train['Age'] < 18].index, inplace=True)

chexpert_train
```

## Sorting for binary classification using only AP Images

```python
chexpert_AP = chexpert_train[chexpert_train['AP/PA'] == 'AP']
chexpert_AP
```

```python
for path in chexpert_AP.index:
    source_path = os.path.join(PATH_TO_CHEXPERT_DOWNLOADS, chexpert_AP.loc[path]['Path'])

    if chexpert_AP.loc[path]['No Finding'] == 1.0:
        target_path = os.rename(source_path, os.path.join(PATH_TO_NORMAL_TRAIN_DIRECTORY_AP, str(chexpert_AP.loc[path]['Path']).replace('/', '_')))
    else:
        target_path = os.rename(source_path, os.path.join(PATH_TO_ABNORMAL_TRAIN_DIRECTORY_AP, str(chexpert_AP.loc[path]['Path']).replace('/', '_')))
```
