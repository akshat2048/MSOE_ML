```python
import pandas
import os
```

```python
PATH_TO_CHEXPERT_DOWNLOADS = ""
PATH_TO_NORMAL_TRAIN_DIRECTORY = ""
PATH_TO_ABNORMAL_TRAIN_DIRECTORY = ""
```

```python
# read in chexpert_train.csv
chexpert_train = pandas.read_csv('chexpert_train.csv')

chexpert_train
```

```python
# remove people with support devices

chexpert_train.drop(chexpert_train[chexpert_train['Support Devices'] == 1].index, inplace=True)

chexpert_train
```

```python
# remove children (under 18)
chexpert_train.drop(chexpert_train[chexpert_train['Age'] < 18].index, inplace=True)

chexpert_train
```

## Sorting for binary classification using only AP Images

```python
chexpert_AP = chexpert_train[chexpert_train['AP/PA'] == 'AP']

chexpert_AP
```

```python
for path in chexpert_AP.index:
    source_path = os.path.join(PATH_TO_CHEXPERT_DOWNLOADS, chexpert_AP.loc[path]['Path'])

    if chexpert_AP.loc[path]['No Finding'] == 1.0:
        target_path = os.rename(source_path, os.path.join(PATH_TO_NORMAL_TRAIN_DIRECTORY, str(chexpert_AP.loc[path]['Path']).replace('/', '_')))
    else:
        target_path = os.rename(source_path, os.path.join(PATH_TO_ABNORMAL_TRAIN_DIRECTORY, str(chexpert_AP.loc[path]['Path']).replace('/', '_')))
    
```
