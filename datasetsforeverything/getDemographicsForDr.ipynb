```python
import os
import pandas as pd
```

```python
nih = pd.read_csv("NIH.csv")
nih.tail()
```

```python
mimic = pd.read_csv("mimic-cxr-2.0.0-metadata.csv")
mimic
```

```python
chexpert = pd.read_csv("train.csv")
chexpert
```

```python
false_positives_files = "/Users/akshatchannashetti/Downloads/images/False Positives"
false_negatives_files = "/Users/akshatchannashetti/Downloads/images/False Negatives"
```

```python
os.listdir(false_positives_files)
os.listdir(false_negatives_files)
```

```python
new_nih = nih.loc[nih['Image Index'].isin(os.listdir(false_negatives_files)) | nih['Image Index'].isin(os.listdir(false_positives_files))]
new_nih.to_csv("nih_demographics.csv")
```

```python
list_of_dicom_ids = []
for file in os.listdir(false_positives_files):
    list_of_dicom_ids.append(file[:-4])
for file in os.listdir(false_negatives_files):
    list_of_dicom_ids.append(file[:-4])

new_mimic = mimic.loc[mimic['dicom_id'].isin(list_of_dicom_ids)]
new_mimic.to_csv("mimic_demographics.csv")
```

```python
chexpert_files = []

for file in os.listdir(false_positives_files):
    if("study" in file):
        chexpert_files.append(file[:-4])

for file in os.listdir(false_negatives_files):
    if("study" in file):
        chexpert_files.append(file[:-4])

tmp = []

for file in chexpert_files:
    tmp.append("CheXpert-v1.0-small/train/" + str(file).replace("_", "/"))

chexpert_files = tmp
tmp = []

for file in chexpert_files:
    tmp.append(file[::-1].replace("/", "_", 1)[::-1] + ".jpg")

chexpert_files = tmp
chexpert_files

new_chexpert = chexpert.loc[chexpert['Path'].isin(chexpert_files)]
new_chexpert.to_csv("chexpert_demographics.csv")
```
