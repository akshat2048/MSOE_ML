```python
# import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
xrays = pd.read_csv('NIH.csv')
```

```python
xrays.drop(xrays[xrays['Finding Labels'].str.contains('\|')].index, inplace=True)

fig, axes = plt.subplots()
fig.set_size_inches(24, 8.27)

sns.countplot(data=xrays, x='Finding Labels', ax=axes)
```
